```python
import datetime
import numpy as np
import pandas as pd

df = pd.DataFrame({'Quarter':['2015Q1','2015Q2']})
df['date'] = pd.to_datetime(df.Quarter).dt.to_period('Q')

df['year'] = df['date'].dt.year
df['quarter'] = df['date'].dt.quarter

print (df)
```
